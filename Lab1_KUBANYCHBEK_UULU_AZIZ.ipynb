<h4>4.	Напишите функцию, которая производит вычисление по формуле</h4>

```python
# Определяем функцию f, которая принимает аргументы x и t
def f(x, t):
    # Вычисляем числитель
    numerator = 9 * t + 10 * math.cos(x)
    
    # Вычисляем знаменатель
    denominator = math.sqrt(t - math.sin(x))
    
    # Вычисляем экспоненциальную часть
    exponent = math.exp(-2 * t)
    
    # Вычисляем результат
    result = (numerator / denominator) * exponent
    
    return result
x = 3 
t = 3  
print(f(x, t))

```

<h4>5.	Напишите функцию, которая из трех чисел находит все принадлежащие интервалу от a до b (границы интервала задать произвольно).</h4>

```python
def find_in_interval(x1, x2, x3, a, b):
    numbers = [x1, x2, x3]
    
    # Находим числа принадлежащие интервалу от а до b
    interval = [x for x in numbers if a <= x <= b]
    
    return interval

# Использование нашей функции
x1, x2, x3 = 15, 9, 6
a, b = 2, 12
result = find_in_interval(x1, x2, x3, a, b)
print(result)  
```

<h4>6.	Создайте генератор 20-ти случайных чисел в диапазоне от 0 до 10.</h4>

```python
import random
#count - Количество случайных чисел для генерации
#lower_bound - Нижняя граница диапазона
#upper_bound - Верхняя граница диапазона

def generate_random_numbers(count, lower_bound, upper_bound):
    for _ in range(count):
        yield random.randint(lower_bound, upper_bound)

# Использование генератора
random_numbers_generator = generate_random_numbers(20, 0, 10)
random_numbers = list(random_numbers_generator)
print(random_numbers)
```

<h4>7.	Вычислите сумму 20 элементов последовательности</h4>

```python
def sequence_sum(n):
    total_sum = 0  # Переменная для хранения суммы
    for i in range(1, n+1):
        term = i / ((i + 1) ** i)  # Вычисляем i-ый элемент последовательности
        total_sum += term  # Добавляем элемент к общей сумме
    return total_sum

# Вычисляем сумму элементов
n = 20
result = sequence_sum(n)
print(f"Сумма первых {n} элементов последовательности: {result}")
```

<h4>6. Оператор `if … else` используется для выполнения различных блоков кода в зависимости от выполнения условия. Его основная задача — управлять потоком выполнения программы на основе условий.
if — проверяет, истинно ли заданное условие.
else — выполняет альтернативный блок кода, если условие в if не выполнено.
Этот оператор позволяет принимать решения в коде и выполнять разные действия в зависимости от того, выполняется ли определённое условие или нет.</h4>








<h4>7. Выполнение оператора `while` прекращается, когда условие в его заголовке становится ложным.
Оператор while продолжает выполнение блока кода, связанного с ним, до тех пор, пока условие истинно. Когда условие становится ложным, цикл завершается, и программа переходит к коду, находящемуся после цикла.

Основные случаи завершения цикла while:

Условие становится ложным — как только условие в заголовке цикла перестает быть истинным, выполнение цикла прекращается.
Оператор break — оператор break можно использовать внутри цикла while, чтобы немедленно завершить выполнение цикла, независимо от условия.
Внешние прерывания — такие как завершение программы или возникновение исключений, также могут прервать выполнение цикла.
Важно следить за тем, чтобы условие цикла в конечном итоге становилось ложным, чтобы избежать бесконечного цикла.</h4>

<h4>8. В Python функции можно создавать следующими способами:

- **Обычные функции** — определяются с помощью ключевого слова `def`, могут принимать аргументы и возвращать значения.

- **Анонимные функции (lambda-функции)** — короткие функции, создаваемые с помощью ключевого слова `lambda`, обычно используются для небольших и простых операций.

- **Функции с параметрами по умолчанию** — функции, в которых некоторые параметры имеют значения по умолчанию, если они не передаются при вызове.

- **Функции с переменным числом аргументов** — функции, способные принимать произвольное количество позиционных и именованных аргументов с использованием `*args` и `**kwargs`.

- **Функции-генераторы** — функции, которые возвращают итератор и используют оператор `yield` для генерации значений по мере их запроса.

- **Функции-декораторы** — функции, принимающие другую функцию в качестве аргумента и возвращающие её модифицированную версию, добавляя дополнительное поведение.</h4>

<h4>9. Лямбда-функция в Python — это анонимная функция, которая создается с помощью ключевого слова `lambda`. Она позволяет определять небольшие функции в одну строку, без необходимости использовать ключевое слово `def` и задавать имя функции.

Основные характеристики лямбда-функций:

- **Анонимность**: Лямбда-функции не имеют имени.
- **Однострочность**: Они предназначены для написания коротких функций, обычно для одноразового использования.
- **Ограничения**: Лямбда-функции могут содержать только одно выражение, результат которого возвращается. Они не поддерживают многострочные инструкции или сложные блоки кода.

Лямбда-функции часто используются в ситуациях, когда функция передается как аргумент, например, в функциях высшего порядка, таких как `map`, `filter`, или `sorted`.</h4>
